# ***DATA Cleaning***

```python
# dataset.isnull().sum()
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
```

```python
dataset.info()
```

```python
dataset[['Region']] = dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
## drop, removing the null values
dataset = dataset.dropna().reset_index(drop=True)
dataset.head()
```

```python
# dataset.isnull()
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns
```

```python
##fix spaces and column names
dataset.columns = dataset.columns.str.strip()
dataset.columns
# dataset.columns = dataset.columns.str.replace(' ', '_')
```

# ***changing the other columns to integer data type***

```python
## chaange the reqd col to int data type
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

# ***Changing the other columns to float data datatype***

```python
# changing the other columns to float data type
# dataset[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = dataset[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
objects = [features for features in dataset.columns if dataset[features].dtypes == 'O']
```

```python
for i in objects:
  if i != 'Classes':
    dataset[i] = dataset[i].astype(float)
```
