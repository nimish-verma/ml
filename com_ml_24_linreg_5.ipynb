# ***Handling Imbalanced Dataset***

Upsampling

```python
import numpy as np
import pandas as pd

np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
#create dataframe w imbalanced dataset
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc = 0 , scale = 1 , size = n_class_0),
    'feature_2': np.random.normal(loc = 0 , scale = 1 , size = n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

**Up sampling**

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, replace = True, n_samples = len(df_majority), random_state=142)
```

```python
df_minority_upsampled
df_minority_upsampled.shape
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
```

**DOWN SAMPLING**

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]
```

```python
from sklearn.utils import resample
df_majority_downsampled = resample(df_minority,
                                   replace=True, #Sample With replacement
                                   n_samples=len(df_minority),
                                   random_state=42)
```

```python
print(df_majority_downsampled.shape)

df_downsampled = pd.concat([df_minority , df_majority_downsampled])
```

```python
df_downsampled.target.value_counts()
```
